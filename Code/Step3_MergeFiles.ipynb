```python
import pandas as pd

# === INPUT FILES ===
ready_file = "comments_ready.csv"
sentiment_file = "comments_weaklabel.csv"
topic_file = "comments_with_topics.csv"

# === LOAD ===
df_ready = pd.read_csv(ready_file, dtype=str, low_memory=False)
df_sent = pd.read_csv(sentiment_file, dtype=str, low_memory=False)
df_topic = pd.read_csv(topic_file, dtype=str, low_memory=False)

# === ENSURE KEY COLUMNS OCCUR ===
for col in ["video_id", "text_clean"]:
    if col not in df_ready.columns or col not in df_sent.columns:
        raise KeyError(f"Thi·∫øu c·ªôt '{col}' ƒë·ªÉ merge ‚Äî h√£y ki·ªÉm tra file!")

# === MERGE STEP ===
df = df_ready.merge(
    df_sent[["video_id", "text_clean", "weak_label"]],
    on=["video_id", "text_clean"],
    how="left"
).merge(
    df_topic[["text_clean", "topic_id"]],
    on="text_clean",
    how="left"
)

# === SAVE MASTER ===
df.to_csv("analysis_master.csv", index=False, encoding="utf-8-sig")
print("‚úÖ Saved analysis_master.csv")
print(df.head())
```

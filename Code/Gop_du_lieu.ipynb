```python
# =====================
# 1. LOAD DATA
# =====================
import pandas as pd

df1 = pd.read_csv('vietnamworks_jobs_detail.csv')
df2 = pd.read_excel('TopCV_Data.xlsx')
df3 = pd.read_excel('data joboko.xlsx')
df4 = pd.read_excel('data linkedin (analyts,sience,machine learining)_CO_ANH.xlsx')
df5 =pd.read_excel('data wed It viec.xlsx')
```

```python
print(df1.columns)
print(df2.columns)
print(df3.columns)
print(df4.columns)
print(df5.columns)
```

```python
df1 = df1.rename(columns={
    'job_category':'Chuyên môn',
    'job_title':'Tiêu đề',
    'company':'Tên công ty',
    'location':'Địa chỉ',
    'date_posted':'Thời gian đăng',
    'skills':'Kỹ năng',
    'salary':'Lương',
    'job_description':'Mô tả công việc',
    'requirements':'Yêu cầu',
    'company_field':'Lĩnh vực',
    'url':'URL'
})

df2 = df2.rename(columns={
    'Tên công việc':'Tiêu đề',
    'Công ty':'Tên công ty',
    'Địa điểm làm việc':'Địa chỉ',
    'Mức lương':'Lương',
    'Kinh nghiệm':'Kinh nghiệm',
    'Mô tả công việc':'Mô tả công việc',
    'Yêu cầu ứng viên':'Yêu cầu',
    'Quyền lợi':'Quyền lợi',
    'Lĩnh vực':'Lĩnh vực',
    'Thời gian làm việc':'Thời gian đăng',
    'URL':'URL'
})

df3 = df3.rename(columns={
    'Tiêu đề':'Tiêu đề',
    'Tên công ty':'Tên công ty',
    'Nơi làm việc':'Địa chỉ',
    'Thu nhập':'Lương',
    'Kinh nghiệm':'Kinh nghiệm',
    'Mô tả':'Mô tả công việc',
    'Yêu cầu':'Yêu cầu',
    'Quyền lợi':'Quyền lợi',
    'Loại hình':'Hình thức',
    'Chức vụ':'Chuyên môn'
})

df4 = df4.rename(columns={
    'Tiêu đề':'Tiêu đề',
    'Công ty':'Tên công ty',
    'Địa chỉ':'Địa chỉ',
    'Ngày đăng':'Thời gian đăng',
    'Mô tả công việc':'Mô tả công việc'
})

df5 = df5.rename(columns={
    'Tiêu đề Job':'Tiêu đề',
    'Tên công ty':'Tên công ty',
    'Địa chỉ':'Địa chỉ',
    'Lương':'Lương',
    'Hình thức':'Hình thức',
    'Thời gian đăng':'Thời gian đăng',
    'Kỹ năng':'Kỹ năng',
    'Chuyên môn':'Chuyên môn',
    'Lĩnh vực':'Lĩnh vực',
    'Yêu cầu (KN)':'Yêu cầu',
    'Miêu tả CV':'Mô tả công việc'
})

# =====================
# 3. THÊM CỘT SOURCE
# =====================
df1['Source'] = 'VietnamWorks'
df2['Source'] = 'TopCV'
df3['Source'] = 'JobOKO'
df4['Source'] = 'LinkedIn'
df5['Source'] = 'ITviec'

# =====================
# 4. CHỌN DATAFRAME CÓ NHIỀU CỘT NHẤT LÀM CHUẨN
# =====================
dfs = [df1, df2, df3, df4, df5]
max_cols_df = max(dfs, key=lambda x: x.shape[1])
common_columns = max_cols_df.columns

# =====================
# 5. ALIGN CỘT – THIẾU THÌ NaN
# =====================
dfs_aligned = [
    df.reindex(columns=common_columns)
    for df in dfs
]

# =====================
# 6. GỘP DATA
# =====================
df_all = pd.concat(dfs_aligned, ignore_index=True)
```

```python
# =====================
# 7. KIỂM TRA
# =====================
print("Số dòng:", df_all.shape[0])
print("Số cột:", df_all.shape[1])
print("\nDanh sách cột:")
print(df_all.columns)

print("\n5 dòng đầu:")
print(df_all.head())
```

```python
df_all.to_excel(
    'data_science_jobs.xlsx',
    sheet_name='All_Jobs',
    index=False
)
```
